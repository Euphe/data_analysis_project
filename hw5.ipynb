```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import scipy
df = pd.read_csv('data/filmdeathcounts.csv', sep=',')
```

## Choosing features

We chose features *Length_Minutes* and *IMDB_Rating*

```python
df.plot.scatter(x='Length_Minutes', y='IMDB_Rating')
```

## Making classificator

```python
training_x = df['Length_Minutes'].values.reshape(-1, 1)
training_y = df['IMDB_Rating'].values
```

```python
classificator = LinearRegression().fit(training_x, training_y)
```

```python
predicted_y = classificator.predict(training_x)
```

```python
plt.scatter(training_x, training_y)
plt.plot(training_x, predicted_y, color='black', linewidth=3)
```

## Properties of regression

Slope:

```python
classificator.coef_
```

Meaning of the slope: every minute of the film gives additional 0.02114397 to the rating

Determinacy coefficient and mean absolute error:

```python
classificator.score(training_x, training_y)
```

```python
mean_absolute_error(training_y, predicted_y)
```

Determinacy coefficient is 0.18600394479373705, it's very low value. Mean absolute error is very high. It shows that classificator works very bad on these features

## Prediction

Trying to predict values

```python
lengths_to_predict = np.array([60, 90, 150])
prdicted_ratings = classificator.predict(lengths_to_predict.reshape(-1, 1))
```

```python
plt.scatter(lengths_to_predict, prdicted_ratings)
```
